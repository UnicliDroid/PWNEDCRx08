```python
# import libraries + wardriving data

import pandas as pd
wd = pd.read_csv ('log11.csv', sep=';', encoding='latin1') # import list of all WiFi devices from CSV file to pandas data frame
wd.sample(10) # sample 10 random devices from dataset
#wd2 = wd.sample(n=100).reset_index(drop=True)
#len(wd) # number of detected devices
```

```python
# start mapping devices

import folium
from folium import Map, Marker
basic_map = folium.Map(location=[9.924659, -84.056675],zoom_start=15) # manually pick scope of map
#folium.PolyLine(wd[['CurrentLatitude', 'CurrentLongitude']].values.tolist(),line_opacity = 0.5, weight = 4).add_to(basic_map) # plot route


for i in range(len(wd)):
    lat = wd.iloc[i]["CurrentLatitude"]
    lon = wd.iloc[i]["CurrentLongitude"]
    ssid = wd.iloc[i]["SSID"]
    folium.Marker(
        location=[lat, lon],
        popup=ssid,
        icon=folium.Icon(icon="wifi", prefix='fa')
    ).add_to(basic_map)


# add markers for every point in dataset
#for i in range(len(wd)):
  # if wd["FirstSeen"][i] > "2021-8-21 18:43:25" and wd["FirstSeen"][i] < "2021-8-21 18:46:58": # add time filter
#  lat = wd["CurrentLatitude"][i]
#  lon = wd["CurrentLongitude"][i]
#  folium.Marker( location=[lat, lon], popup=wd["SSID"][i], icon=folium.Icon(icon="wifi", prefix='fa') ).add_to(basic_map) # annotate GPS location w/ SSID

display(basic_map)
```
